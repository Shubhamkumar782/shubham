```python
import re
import os
import pathlib
from pathlib import Path, PurePath
import pandas as pd
import numpy as np
import logging
from openpyxl.workbook import Workbook

 

keywords = ["prot","prot n","Id iniziativa","PG","Protocollo","Protocollo n","Registro comunicazione","Registro di sistema num"]

 


def Numero(keywords,raw_text):
    ls = {}
    for keyword in keywords:
#         reg_PI = re.compile(r"\b%s.(\d{4,11})"%re.escape(keyword))
        ex_PI = re.findall(r"(?<=%s)\s{0,4}.{1,10}?\s{0,4}(\d{3,11})"%re.escape(keyword), raw_text, flags=re.I)
        if len(ex_PI):
            if int(ex_PI[0])>100000:
                print(keyword,'--',ex_PI)
                ls[keyword]=int(ex_PI[0])
                break
            continue
        continue
    return ls
path = Path(r"D:\Users\SKuma557\OneDrive - JNJ\Desktop\Bnt project\text_files")

 


def main_Numero(keywords, path):
    os.chdir(Path(path))
    main_dic = {}
    for file in os.listdir(Path(path)):
        print(file)
        try:
            with open(file, 'r') as textfile:
                try:
                    #                 raw_text = textfile.read()
                    lines = textfile.readlines()
                    newstr = ' '.join([line.strip() for line in lines]).replace('/','').replace('-','').replace('.','')

 

                    ret = Numero(keywords, newstr)
                    print(ret)
                    main_dic[file] = ret
                except Exception as e:
                    print(e)
                    logging.critical(f" could not process the file {file} , Error = {e}")
                    logging.debug(e)
                    continue
        except Exception as e:
            print(e)
            logging.critical(f" could not read the file {file} , Error = {e}")
            logging.debug(e)
            continue

 

    return main_dic

 

main_dic = main_Numero(keywords, path)
main_dic

 

df = pd.DataFrame(data = main_dic).T
os.chdir(r"D:\Users\SKuma557\OneDrive - JNJ\Desktop\Bnt project")
df.to_excel("Numero extraction updated_final_.xlsx")
```
